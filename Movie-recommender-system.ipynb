```python
import numpy as np
import pandas as pd
import ast 
```

# Data Cleaning 

*Items to be kept(Keeping in mind that we have to create tags)*
1. Generes
2. ID
3. Keywords
4. title(remove original title)
5. Overview
6. cast
7. crew

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#we have to use literal_eval function from ast module as the genres are in the form of string and has to be converted to list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#doing the same step for keywords 
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#taking out only important characters from cast 
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew']
```

```python
def fetch_director(obj):
    if obj is None or obj == []:
        return []  # Return an empty list if obj is None or empty
        
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies['tags'][0]
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
vectors
```

```python
vectors[0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#calculating the cosine distance(angular) between vectors
#because euclidian distance is not a good measure for the high dimensional data (here 5000)
# hence use cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#distance of each movie with one another
cosine_similarity(vectors) 
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6] # value of similarity with itself is 1
#diagonal elements of this similarity matrix will be 1
```

```python
sorted(similarity[0],reverse=True)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie ].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
